# First file

```python
df1 = pd.read_csv("v0_original-from-docs.csv")
df1 = df1.replace('\t', ' ', regex=True)

df1.head()
```

```python
import re

regex = r"^(\d+\.)?(.+)\s(\S+ lt.)\s+(€.+)"
df1['external_id'] = df1['full_data'].str.extract(regex)[0]
df1['name'] = df1['full_data'].str.extract(regex)[1]
df1['size'] = df1['full_data'].str.extract(regex)[2]
df1['price'] = df1['full_data'].str.extract(regex)[3]

name_regex = r"(.+)(\d{4})(.+)"

# if matches, then extract vintage
for i, row in df1.iterrows():
    full_name = row['name']
    if isinstance(full_name, str) and re.match(name_regex, full_name):
        match = re.match(name_regex, full_name)
        df1.at[i, 'vintage'] = match.group(2)
        df1.at[i, 'winery'] = match.group(3)
        df1.at[i, 'name'] = match.group(1)
    else:
        df1.at[i, 'vintage'] = None
        df1.at[i, 'winery'] = None

df1.head(25)
```

```python
# df to csv
# delete full_data column
df1 = df1.drop(columns=['full_data'])
df1.to_csv("v0_preprocessed.csv", index=False)
```
